# Cost Function for Reccomender System (simple implementation)

```python
def costWithRegCoFi(X, theta, Y, R, lambda_):
  #n_m, n_u = Y.shape
  n_m = 5
  n_u = 4
  n = 3
  #n = X.shape[1]
  cost = 0
  regX = 0
  regtheta = 0
  for i in range(n_m):
    for j in range(n_u):
      if(R[i, j]==1):
        cost = cost + (1/2) * (np.dot(theta[j,:n].T, X[i,:n]) - Y[i, j])**2

  for i in range(n_m):
    for k in range(n):
      regX = regX + X[i, k]**2

  for j in range(n_u):
    for k in range(n):
      regtheta = regtheta + theta[j, k]**2

  costWithReg = cost +(lambda_/2) * (regX + regtheta)

  return costWithReg    
```

```python
cost = costWithRegCoFi(X, Theta, Y, R, lambda_=0)
```

# Gradient Descent for Reccomender System (simple implementation)


<p align="center" width="100%">
    <img width="100%" src="https://drive.google.com/uc?id=1PB5CmuxDP1yI-Fd68sfq1YYPweaDLmsk">
</p>

```python
def gradientDescentCoFi(X, theta, Y, R, lr, epochs, lambda_):
  n_m, n_u = Y.shape
  n = X.shape[1]


  for epoch in range(epochs):
    #print(epoch)
    gradx = 0
    gradtheta = 0  
    for k in range(n):
      gradx = 0
      for i in range(n_m):
        for j in range(n_u):
          if(R[i, j]==1):
            gradx = gradx + (np.dot(theta[j, :].T, X[i,:]) - Y[i, j]) * theta[j, k] 

        X[i, k] = X[i, k] - lr * (gradx + lambda_ * X[i, k])

    for k in range(n):
      gradtheta = 0
      for j in range(n_u):
        for i in range(n_m):
          if(R[i, j]==1):
            gradtheta = gradtheta + (np.dot(theta[j, :].T, X[i,:]) - Y[i, j]) * X[i, k] 

        theta[j, k] = theta[j, k] - lr * (gradtheta + lambda_ * theta[j, k])

  return X, theta        
```

```python
lr = 0.01
lambda_ = 1
epochs = 2

Xnew, thetanew = gradientDescentCoFi(X, Theta, Y, R, lr, epochs, lambda_)
```

# **mean normalization**

<p align="center" width="100%">
    <img width="100%" src="https://drive.google.com/uc?id=1dhlOH8fbslUUY8bTYJEUxlOmVXmb8zg9">
</p>

```python
def normalizeRatings(Y, R):
  n_m, n_u = Y.shape
  Ymean = np.zeros(n_m)
  Ynorm = np.zeros(Y.shape)

  for i in range(n_m):
    idx = (R[i, :] == 1)
    Ymean[i] = np.mean(Y[i, idx])
    Ynorm[i, idx] = Y[i, idx] - Ymean[i]

  return Ynorm, Ymean
```

```python
Ynorm, Ymean = normalizeRatings(Y, R)
```

```python
lr = 0.001
lambda_ = 1
epochs = 10000
#n_m, n_u = Y.shape
#n = 10
n_m = 10
n_u = 10
n = 4
#initialize X and Theta
X1 = np.random.randn(n_m, n)
Theta1 = np.random.randn(n_u, n)
Xnew, thetanew = gradientDescentCoFi(X1[:n_m, :n], Theta1[:n_u, :n], Ynorm[:n_m, :n_u], R[:n_m, :n_u], lr, epochs, lambda_)
```

```python
p = np.dot(Xnew, thetanew.T)
my_pred = p[:,0] + Ymean[:n_m]
print(my_pred)
ix = np.argsort(my_pred)[::-1]
print(ix)
print('Top recommendations for new user')
for i in range(10):
    j = ix[i]
    print('Predicting rating %.1f for movie %s' % (my_pred[j], movieList[j]))

print('\nOriginal ratings provided:')

for i in range(len(my_ratings)):
    if my_ratings[i] > 0:
        print('Rated %d for %s' % (my_ratings[i], movieList[i]))
```
