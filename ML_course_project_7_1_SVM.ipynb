```python
#load data set from project7data1.mat file
#ذخیره شده است .mat دیتاست این پروژه در یک فایل با فرمت 
#first upload project6data1.mat to colab
#ابتدا فایل مربوطه را در این محیط آپلود کرده
#load the .mat file using scipy library
#تصاویر را در یک متغیر لود می کنیم scipy با استفاده از کتابخانه 
import numpy as np
from scipy.io import loadmat

data = loadmat('/content/project7data1.mat')

```

```python
#split the data into X, Y
#داده ها ی ورودی و خروجی را در دو متغیر جدا میریزیم
#training input
X = data['X']
#training output
Y = data['y']

#find m
m = Y.shape[0]
print(m)
```

# Data Visualization:

# نمایش داده ها

```python
import matplotlib.pyplot as plt
# Find Indices of Positive and Negative Examples
pos = Y[:, 0] == 1
neg = Y[:, 0] == 0
print(pos)
print(neg)
   
plt.plot(X[pos, 0], X[pos, 1], 'X', mew=1, ms=10, mec='k')
plt.plot(X[neg, 0], X[neg, 1], 'o', mew=1, mfc='y', ms=10, mec='k')
plt.grid(True)
```

```python
#use sklearn for SVM
from sklearn import svm
#create model
clf = svm.SVC(kernel='linear', C=100)
#train model
clf.fit(X, Y[:,0])
#
```

```python
w = clf.coef_[0]           # w consists of 2 elements
b = clf.intercept_[0]      # b consists of 1 element
x_points = np.linspace(0,5)    # generating x-points from -1 to 1
y_points = -(w[0] / w[1]) * x_points - b / w[1]


plt.plot(X[pos, 0], X[pos, 1], 'X', mew=1, ms=10, mec='k')
plt.plot(X[neg, 0], X[neg, 1], 'o', mew=1, mfc='y', ms=10, mec='k')


plt.plot(x_points, y_points, c='r');
```

```python
#load data set from project7data2.mat file
#ذخیره شده است .mat دیتاست این پروژه در یک فایل با فرمت 
#first upload project7data2.mat to colab
#ابتدا فایل مربوطه را در این محیط آپلود کرده
#load the .mat file using scipy library
#تصاویر را در یک متغیر لود می کنیم scipy با استفاده از کتابخانه 
import numpy as np
from scipy.io import loadmat

data = loadmat('/content/project7data2.mat')

```

```python
#split the data into X, Y
#داده ها ی ورودی و خروجی را در دو متغیر جدا میریزیم
#training input
X = data['X']
#training output
Y = data['y']



#find m
m = Y.shape[0]
print(m)
```

```python
import matplotlib.pyplot as plt
# Find Indices of Positive and Negative Examples
pos = Y[:, 0] == 1
neg = Y[:, 0] == 0
   
plt.plot(X[pos, 0], X[pos, 1], 'X', mew=1, ms=10, mec='k')
plt.plot(X[neg, 0], X[neg, 1], 'o', mew=1, mfc='y', ms=10, mec='k')
plt.grid(True)
```

```python
#use sklearn for SVM
from sklearn import svm
#create model
clf = svm.SVC(kernel='rbf', C=1000, gamma=100)
#train model
clf.fit(X, Y[:,0])
#
```

```python
# create a mesh to plot in
#step size in mesh
h=0.02

x_min, x_max = X[:, 0].min() , X[:, 0].max()
y_min, y_max = X[:, 1].min() , X[:, 1].max()
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)

# Plot also the training points
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.coolwarm)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())



plt.show()
```

```python
#load data set from project7data3.mat file
#ذخیره شده است .mat دیتاست این پروژه در یک فایل با فرمت 
#first upload project7data3.mat to colab
#ابتدا فایل مربوطه را در این محیط آپلود کرده
#load the .mat file using scipy library
#تصاویر را در یک متغیر لود می کنیم scipy با استفاده از کتابخانه 
import numpy as np
from scipy.io import loadmat

data = loadmat('/content/project7data3.mat')

```

```python
data.keys()
```

```python
#split the data into X, Y
#داده ها ی ورودی و خروجی را در دو متغیر جدا میریزیم
#training input
X = data['X']
#training output
Y = data['y']

#cross validation input
Xval = data['Xval']
#cross validation output
Yval = data['yval']

#find m
m = Y.shape[0]
print(m)
```

```python
import matplotlib.pyplot as plt
# Find Indices of Positive and Negative Examples
pos = Y[:, 0] == 1
neg = Y[:, 0] == 0
   
plt.plot(X[pos, 0], X[pos, 1], 'X', mew=1, ms=10, mec='k')
plt.plot(X[neg, 0], X[neg, 1], 'o', mew=1, mfc='y', ms=10, mec='k')
plt.grid(True)
```

```python
#use sklearn for SVM
from sklearn import svm


gamma_selection=[0.1, 1, 10, 100, 1000]
C_selection = [1, 10, 100, 1000]
#create model
accuracy=[]
for gamma_ in gamma_selection:
  for C_ in C_selection:
    clf = svm.SVC(kernel='rbf', C=C_, gamma=gamma_)
    #train model
    clf.fit(X, Y[:,0])
    acc=clf.score(Xval, Yval)
    accuracy.append(acc)
    print("accuracy of model for C=%d and gamma=%f is %f" %(C_, gamma_, acc))
```
